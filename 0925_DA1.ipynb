#### READING SPOTIFY DATASET

```python
spotify_df = pd.read_csv(r"Spotify Most Streamed Songs.csv",dtype={'artist_count':'str','track_name':"str",'artist(s)_name':'str','key':'str'})
spotify_df
```

#### EDA (EXPLORATORY DATA ANALYSIS)

##### CHECKING FOR DUPLICATES/REDUNDANCIES IN DATASET

```python
print(spotify_df.shape)
print(spotify_df.drop_duplicates().shape)
```

```python
spotify_df.columns
```

```python
df2 = spotify_df.groupby('released_year')['artist_count'].sum()
```

```python
df2
```

```python
spotify_df['Date'] = spotify_df['released_year'].astype(str)+ '-'+ \
spotify_df['released_month'].astype(str)+ '-'+ \
spotify_df['released_day'].astype(str)
spotify_df['Date']
```

```python
spotify_df['Date'] = pd.to_datetime(spotify_df['Date'],format='%Y-%m-%d')
spotify_df['Date']
```

```python
spotify_df
```

```python
spotify_df['YRMO'] = spotify_df['Date'].dt.strftime('%Y%m')
spotify_df
```
